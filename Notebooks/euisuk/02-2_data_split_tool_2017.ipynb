## 2017년 추가

```
import pandas as pd
import numpy as np
import os, shutil
from glob import glob
from tqdm import tqdm
import time
import logging
```

```
logger = logging.getLogger(__name__)
logging.basicConfig(format='%(asctime)s - %(message)s', level=logging.INFO)

base_dir = '/content/drive/Shareddrives/예측모델_기말프로젝트/원본_데이터/folder_2017'
data_2017 = glob(os.path.join(base_dir, "*.csv"))
data_2017
```

```
# set clean columns
data1 = pd.read_csv(data_2017[0], encoding = 'euc-kr')
clean_columns = list(map(lambda r: r.replace('\'', ''), list(data1.columns)))
```

```
clean_columns
```

```
# ------------------------
#     DATA CLEANSING
# ------------------------
result_df = pd.DataFrame(columns = clean_columns)
for file in tqdm(data_2017, desc = "DATA CONCAT"):
    # read
    to_get = pd.read_csv(file, encoding = 'euc-kr')

    # change columns
    to_get.columns = clean_columns
    
    # change values
    for i in range(7):
        to_get.iloc[:, i] = to_get.iloc[:, i].str.replace('\'', '')
    
    to_get_1308 = to_get.loc[to_get['대여소번호'] == '1308', :]    
    logger.info(f"extracted data length: {to_get_1308.shape[0]}")
    
    result_df = pd.concat([result_df, to_get_1308], axis = 0)
```

```
anam_sorted = result_df.sort_values(by = '대여일자').reset_index(drop = True)
anam_sorted.head(5)
```

```
anam_sorted.loc[:, '이용건수'] = pd.to_numeric(anam_sorted.loc[:, '이용건수'])
```

```
anam_sum = anam_sorted.groupby('대여일자', as_index = False).sum()
anam_fin = anam_sum.loc[:, ['대여일자', '이용건수']]
anam_fin.columns = ['date', 'cnt']
```

```
print(len(anam_fin))
```

## Holiday / 전처리

```
import holidays
kor_holidays = holidays.Korea()
```

```
holiday_list = ['-'.join((str(holi.year), str(holi.month).zfill(2), str(holi.day).zfill(2))) for year in range(2017, 2021) for holi in holidays.KR(years=year)]
```

```
holiday_df = pd.DataFrame({'date': holiday_list, 'is_holiday': 1})
holiday_df.head(5)
```

```
holiday_joined = pd.merge(df_fin, holiday_df, on = 'date', how = 'left').fillna(0).astype({'is_holiday': int})
```

```
holiday_joined
```

```
holiday_joined[['year','month', 'day']] = holiday_joined.date.str.split("-",expand=True,)
holiday_joined['weekday'] = pd.to_datetime(holiday_joined['date']).dt.weekday
holiday_joined['weeknum'] = pd.to_datetime(holiday_joined['date']).dt.week
```
